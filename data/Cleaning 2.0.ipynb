```python
import json
import pandas as pd
from pprint import pprint
from fuzzywuzzy import process
```

```python
africa_iso_code = pd.read_csv('./africa_iso_codes.csv', dtype={'name': str})
africa_iso_code
```

```python
trade_data = pd.read_csv('./cleaned_data_2014.csv', dtype={'commodity_group': str})
trade_data
```

```python
product_names = pd.read_csv('./products_prospedia.csv',dtype={'code': str})
product_names = product_names[['code', 'name', 'level']]
product_names = product_names[product_names['level'] == '2digit']
```

```python
product_names.tail(5)
```

```python
merged_data = pd.merge(trade_data, product_names,
              left_on=['commodity_group'],
              right_on=['code'])
```

```python
merged_data= pd.merge(merged_data, africa_iso_code,
              left_on=['exporter'],
              right_on=['3digit'])
```

```python
merged_data.rename(columns={'name_x': 'product_name', 'name_y': 'country_name'}, inplace=True)
```

```python
len(trade_data)
```

```python
merged_data = merged_data[[
        'commodity_group', 'export_value', 
        'import_value','product_name', 'country_name', 
        '2digit', '3digit']]
```

```python
merged_data.to_csv('./visualized_data.csv',index=False)
```
